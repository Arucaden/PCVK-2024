```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from google.colab.patches import cv2_imshow
```

```python
urls = ["https://i.scdn.co/image/ab67616d0000b273360b42462a1c78b8bb5037d8",
        "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcQKQLaiYJ9gisKnocf7eGMUCNTBK2e2MZDWbw&s",
        "https://assets.jalantikus.com/assets/cache/720/720/userfiles/2023/02/20/NEKO-APK-42f93.jpg"]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  final_frame = cv.hconcat((image, image_2))
  cv2_imshow(final_frame)
  print('\n')
```

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ", tinggi, " x " , lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(lebar):
  image_3[int((tinggi)/2),y] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

**Tugas**

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(tinggi):
  image_4[y, int((lebar)/2)] = [255, 255, 255]

for x in range(min(lebar, tinggi)):
    image_5[x, x] = [255, 253, 253]
    image_5[x, lebar - 1 - x] = [255, 255, 255]

row1 = cv.hconcat((image_2, image_4))
row2 = cv.hconcat((image_2, image_5))

final_frame = cv.vconcat((row1, row2))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_6 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for x in range(20, (int((lebar)/2))):
  image_6[int((tinggi)/2), x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_6))
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_7 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range((int((lebar)/4)), (int((lebar)/2))):
  for x in range((int((lebar)/4)), (int((lebar)/2))):
    image_7[y, x] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_7))
cv2_imshow(final_frame)
```
