```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv

img = cv.imread('/content/drive/MyDrive/WhatsApp Image 2024-06-20 at 12.04.49.jpeg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# 1. Dengan menggunakan figsize, perhatikan apakah ukuran image pixelnya juga berubah
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.title('Figsize')
plt.show()

# 2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja
rb_img = img_rgb.copy()
rb_img[:, :, 1] = 0
plt.imshow(rb_img)
plt.title('Red-Blue Channel')
plt.show()

gb_img = img_rgb.copy()
gb_img[:, :, 0] = 0
plt.imshow(gb_img)
plt.title('Green-Blue Channel')
plt.show()

# 3. Tampilkan image baris ke 20-115, kolom 25-120
cropped_img = img_rgb[20:115, 25:120]
plt.imshow(cropped_img)
plt.title('Cropped Image')
plt.show()

# 4. Tampilkan image baris ke 5-30, semua kolom, channel Red saja
red_img = img_rgb[5:30, :, 0]
plt.imshow(red_img, cmap='Reds')
plt.title('Cropped with Red Channel')
plt.show()

# 5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image
height, width, x = img_rgb.shape
ktk_img = img_rgb.copy()

for x in range(5):
    top_left = (np.random.randint(0, width-50), np.random.randint(0, height-50))
    bottom_right = (top_left[0] + np.random.randint(30, 100), top_left[1] + np.random.randint(30, 100))
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    cv.rectangle(ktk_img, top_left, bottom_right, color, -1)  # -1 untuk mengisi kotak

plt.imshow(ktk_img)
plt.title('Random Rectangles')
plt.show()

# 6. Tampilkan image dengan posisi terbalik
flip_img = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)
plt.imshow(flip_img)
plt.title('Flipped Image')
plt.show()
```

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('/content/drive/MyDrive/WhatsApp Image 2024-06-20 at 12.04.49.jpeg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

blue_channel = img_rgb[:, :, 2]

out_img = blue_channel.copy()

height, width = out_img.shape
rect_width = 100
rect_height = 100

# Koordinat
bottom_left = (0, height - rect_height)  # (x, y)
top_right = (rect_width, height)  # (x, y)

#Soal no.8
# Gambar rectangle pada channel Blue
out_img[bottom_left[1]:top_right[1], bottom_left[0]:top_right[0]] = 255

#Soal no.9
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(out_img, text='This is Blue Channel', org=(170, 880), fontFace=font, fontScale=2, color=(255, 255, 0), thickness=8, lineType = cv.LINE_AA)

# Tampilkan hasil
plt.imshow(out_img, cmap='Blues')
plt.title('Blue Channel with Rectangle')
plt.show()
```
